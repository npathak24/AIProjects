```python
# Extract team1 and team2 values into a single team column
team1_df = df[['team1', 'number of goals team1']].rename(columns={'team1': 'team', 'number of goals team1': 'goals'})
team2_df = df[['team2', 'number of goals team2']].rename(columns={'team2': 'team', 'number of goals team2': 'goals'})
combined_df = pd.concat([team1_df, team2_df])

# Calculate the total goals for each team
goals_by_team = combined_df.groupby('team')['goals'].sum()

# Sort the teams by the total penalties and select the top 10
goals_by_top_10_teams = goals_by_team.nlargest(10)

# Plot the number of goals for each team
plt.figure(figsize=(12, 6))
goals_by_top_10_teams.sort_values(ascending=False).plot(kind='bar')
plt.title('Number of Goals by Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Goals')
plt.xticks(rotation=45)
plt.show()
```

```python
# Combine the team1 and team2 columns into a single team column
df['team'] = df['team1'].fillna('') + ' ' + df['team2'].fillna('')

# Calculate the total penalties for each team
penalties_by_team = df.groupby('team')['penalties scored team1', 'penalties scored team2'].sum()
penalties_by_team['Total Penalties'] = penalties_by_team['penalties scored team1'] + penalties_by_team['penalties scored team2']

# Sort the teams by the total penalties and select the top 10
top_10_teams = penalties_by_team['Total Penalties'].nlargest(10)

# Plot the top 10 teams with the most penalties
plt.figure(figsize=(12, 6))
top_10_teams.plot(kind='bar')
plt.title('Top 10 Matches with Most Penalties Scored')
plt.xlabel('Teams')
plt.ylabel('Total Penalties')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract team1 and team2 values into a single team column
team1_df = df[['team1']].rename(columns={'team1': 'team'})
team2_df = df[['team2']].rename(columns={'team2': 'team'})
combined_df = pd.concat([team1_df, team2_df])

# Count the number of matches played by each team
matches_played_by_team = combined_df['team'].value_counts().nlargest(5)

# Plot a pie chart for the top 5 teams with the most matches played
plt.figure(figsize=(8, 8))
plt.pie(matches_played_by_team, labels=matches_played_by_team.index, autopct='%1.1f%%')
plt.title('Top 5 Teams with Most Matches Played')
plt.show()
```

# Lab -1 Apply Deep Learning Model

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense

# Load the dataset
df = pd.read_csv("C:\Users\naman\OneDrive\Documents\AI Sem2 Assignments\AIDI-2000_Lab-1_StId-100899717/Fifa_world_cup_matches.csv")

# Select input features and output labels
features = ['total attempts team1', 'total attempts team2',
            'conceded team1', 'conceded team2', 'assists team1', 'assists team2',
            'on target attempts team1', 'on target attempts team2', 'off target attempts team1',
            'off target attempts team2', 'attempts inside the penalty area team1', 
            'attempts inside the penalty area  team2', 'attempts outside the penalty area  team1', 
            'attempts outside the penalty area  team2', 'left channel team1', 'left channel team2',
            'left inside channel team1', 'left inside channel team2', 'central channel team1',
            'central channel team2', 'right inside channel team1', 'right inside channel team2',
            'right channel team1', 'right channel team2', 'penalties scored team1', 'penalties scored team2',
           'goal preventions team1', 'goal preventions team2', 'own goals team1',
           'own goals team2', 'forced turnovers team1', 'forced turnovers team2',
           'defensive pressures applied team1', 'defensive pressures applied team2']
labels = ['number of goals team1', 'number of goals team2']

# Prepare the data
X = df[features].values
y = df[labels].values

# Scale the input features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Build the model
model = Sequential()
model.add(Dense(30, input_dim=X.shape[1], activation='sigmoid'))
model.add(Dense(15, activation='sigmoid'))
model.add(Dense(y.shape[1], activation='sigmoid'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)

# Evaluate the model
loss = model.evaluate(X_test, y_test)
print("Mean Squared Error:", loss)
```

# Analysis

We have used Keras (deep learning library) and applied here Sequential model to build a multi layer ANN.                 The number of neurons in the input layer is determined by the number of selected features from the dataset. The second layer has 30 neurons, the third layer has 15 neurons, and the output layer has a number of neurons based on the number of labels. The activation function used for all layers is the sigmoid function. The model is compiled with the mean squared error loss function and the Adam optimizer. It is trained for 100 epochs with a batch size of 32. Finally, the model is evaluated on the test data, and the mean squared error is printed as 0.6815600991249084.
